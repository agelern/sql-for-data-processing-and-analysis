```python
%%sql

SELECT *, ("High Price" - "Low Price") / 2 + "Low Price" AS "Midpoint Price" 
FROM us_pumpkin_prices 
LIMIT 5;
```

```python
%%sql

SELECT * FROM us_pumpkin_prices LIMIT 5;
```

```python
%%sql

SELECT *, "High Price" - "Low Price" AS "Price Range"
FROM us_pumpkin_prices
ORDER BY "Price Range"
LIMIT 5;
```

```python
%%sql

SELECT COUNT(*)
FROM us_pumpkin_prices;
```

```python
%%sql
SELECT MAX("High Price") FROM us_pumpkin_prices;
```

```python
%%sql

SELECT "Variety", COUNT(*)
FROM us_pumpkin_prices 
GROUP BY "Variety"
ORDER BY COUNT DESC;
```

```python
%%sql

SELECT "City Name", "Variety", COUNT(*) 
FROM us_pumpkin_prices 
GROUP BY "City Name", "Variety" 
HAVING MAX("High Price") > 300 
ORDER BY "City Name";
```

```python
%%sql

SELECT "City Name", "Variety"
FROM us_pumpkin_prices
GROUP BY "City Name", "Variety"
HAVING MIN("Low Price") >= 5
ORDER BY "Variety";
```

```python
%%sql

SELECT artist, COUNT(*) AS "hits"
FROM billboard_top_100_year_end
GROUP BY artist
ORDER BY "hits" DESC
LIMIT 10;
```

```python
%%sql

SELECT "City Name", AVG("High Price" - "Low Price") AS "Price Range"
FROM us_pumpkin_prices
GROUP BY "City Name"
ORDER BY "Price Range" DESC;
```
